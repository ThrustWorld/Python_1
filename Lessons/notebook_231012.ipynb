# Istruzione (o enunciato) `if`


L'*istruzione `if`*  un caso particolare di *istruzione composta*, ovvero una istruzione che contiene al suo interno altre istruzioni.



Le parole `if` ed `else` si chiamano *parole chiave*. Non possono essere usate per i nomi di variabile (n per nomi di funzione, metodo, o altro). L'elenco di parole chiave di Python  abbastanza limitato, lo trovate su https://docs.python.org/3/reference/lexical_analysis.html#keywords

Ogni istruzione composta  costituita da una o pi *clausole*. Ogni clausola ha una *intestazione* e una *suite*. Una intestazione inizia con una parola chiave e termina col simbolo di due punti. La suite  composta da una o pi istruzioni correttamente indentate (cio con degli spazi iniziali che la collocano pi a destra delle intestazioni)

# Operatori relazionali

Gli operatori relazioni confrontano due numeri (o due stringhe) tra di loro. Per carenza di simboli sulla tastiera, il modo con cui si scrivono in Python  diverso da quello con cui si scrivono normalmente in matematica.

|Python | Matematica|
|-------|-----------|
|<      | <         |
|<=     |          |
|>      | >         |
|>=     |          |
|==     | =         |
|!=     |          |

Notare in particolare che l'uguaglianza si scrive `==` e non semplicemente `=` perch il simbolo `=`  riservato per l'assegnazione di un varlore a una variabile.

Il risultato di un operatore relazionale  `True` (se la condizione  vera) o `False` (se la condizione  falsa). Vedremo pi in dettaglio cosa sono questi due valori in una delle prossime lezioni.

## Operatori relazionali su numeri

````python
2 > 0
````

````python
2 < 0
````

````python
3 <= 4
````

````python
3 <= 3
````

````python
# Errore: si scrive <= e non =<
3 =< 4
````

````python
# falso, perch 3 non  minore stretto di 3
3 < 3
````

````python
# vero, perch 3  minore o uguale a 3
3 <= 3
````

````python
# l'uguaglianza nelle condizioni si scrive con == e non con =
3 == 5 - 2
````

````python
# questo da errore perch al posto di 3 ci dovrebbe essere il nome di una variabile
3 = 5
````

````python
#  falso perch 1.0/3.0  vicino a 0.333333 ma non uguale
1.0/3.0 == 0.333333
````

````python
# Ecco il valore di 1.0/3.0 . Notare che il valore  sbagliato, il risulato
#  un numero periodico, ma il Python memorizza solo un valore approssimato
1.0/3.0
````

````python
# Posso mischiare numeri interi e in virgola mobile, Python automaticamente
# fa le dovute conversioni.
1 < 3.0
````

## Operatori relazioni su stringhe

Il significato di `==` e `!=`  ovvio. Per gli operatori `<`, `<=`, `>=` e `>` si utilizza l'ordine lessicographico (quello seguito dai dizionari). La descrizione dettagliata di come funziona l'ordine lessicografico  presente sul libro di testo, pagina ....

````python
"ciao" < "zuzzerellone"
````

````python
"tom" < "tomato"
````

Un aspetto particolare  che le lettere maiuscole vengono prima, nell'ordine lessicografico, di tutte le lettere minuscole.

````python
"ciao" < "Zuzzurellone" # falso perch le maiuscole vegono prima delle minuscole
````

````python
"tom" < "Tomato" # falso perch le maiuscole vegono prima delle minuscole
````

Lo spazio viene anche prima delle lettere maiuscole.

````python
" zuzzerellone" < "Ciao" # vero perch lo spazio viene prima di tutte le lettere
````

````python
"car manufacturer" < "carburetor"
````

Non  possibile confrontare numeri e stringhe tra di loro, anche se la stringa contiene solo caratteri numerici.

````python
"2" < 4
````

Se si confrontano due stringhe che contengono numeri, bisogna stare attenti perch l'ordine usato  quello lessicografico, non quello numerico. Questo vuol dire che, siccome `"2" < "8"` (come ovvio), sara `"20" < "8"`.  esattamente lo stesso motivo per cui da `"a" < "p"` segue `"amico" < "p"`.

````python
"20" < "8" # Vero perch  "2" < "8"
````

Dal punto di visto dell'ordinameto, le cifre numeriche vengono prima delle lettere maiuscole ma dopo gli spazi.

````python
"999999999999" < "Amico" # vero, perch le cifre vengono prima delle lettere
````

````python
" " < "1"  # vero, perch lo spazio viene prima di qualunque cifra
````

In generale, l'ordine dei caratteri (spazi < cifre < lettere maiuscole < lettere minuscole) dipende da come sono codificati i caratteri nella memoria del computer. Ogni carattere corrisponde ad un numero, secondo la codifica Unicode. La parte pi importante di Unicode  quella relativa ai caratteri dell'alfabeto latino, cifre numeriche e segni di interpunzione standard, che prende il nome di [codice ASCII](https://www.rapidtables.com/code/text/ascii-table.html). Maggiori dettagli dovrebbe fornire il corso di *Fondamenti di Informatica*.

Il fatto che nel codice ASCII le parentesi quadre vengono prima delle lettere minuscole ma dopo le maiuscole giustifica i seguenti risultati.

````python
"Amico" < "[xyz]" # vero
````

````python
"amico" < "[xyz]" # falso
````
